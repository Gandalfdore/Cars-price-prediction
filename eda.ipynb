* The most expensive brands are: jaguar, mercedes-benz, bmw and porsche

Major conlusion:
* The bigger the car, the bigger engine it needs and the more likely it is to use more advanced and expensive engine with rear-wheel-drive. All of the above, plus the car brand label are major determinants of the price (and correlate pisitively-linearly with the price).

## MCA

```python
import prince
from sklearn.preprocessing import LabelEncoder
```

```python
df
```

```python
# Identify numerical columns
numerical_columns = df.select_dtypes(include='number').columns
df_categorized = df

# Binning numerical columns into named bins
bin_labels = ['Low', 'Medium-Low', 'Medium-High', 'High']

# Binning numerical columns into percentiles
for column in numerical_columns:
    df_categorized[column] = pd.cut(df_categorized[column], bins=4, labels=bin_labels)

df_categorized = df_categorized.astype(object)
```

```python
df_categorized
```

```python
mca = prince.MCA(
    n_components=2,
    n_iter=6,
    copy=True,
    check_input=True,
    engine='auto',
    random_state=42)

mca = mca.fit(df_categorized)

ax = mca.plot_coordinates(
    X=df_categorized,
    ax=None,
    figsize=(50, 50),
    show_row_points=True,
    row_points_size=10,
    show_row_labels=False,
    show_column_points=True,
    column_points_size=30,
    show_column_labels=True,
    legend_n_cols=1)

ax.get_figure().savefig('images/mca_plot.png')
```
